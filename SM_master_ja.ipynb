```python
# coding: utf-8
"""
SubModularを実験するためのコード
"""
from filer2.filer2 import Filer
from library.submodular import SubModular
from library.rouge import Rouge_N
from library.slackbot import SlackBot
import numpy as np
import glob
```

```python
list_test_path = glob.glob('./files/rakuten_ichiba/dumpfile/test/*')
list_ans_path = glob.glob('./files/rakuten_ichiba/dumpfile/ans/*')

list_test_path.sort()
list_ans_path.sort()

print len(list_test_path)
print len(list_ans_path)
```

```python
counter = 1

username = 'プログラムSM'

try:
    for test_path, ans_path in zip(list_test_path, list_ans_path):
        list_test = Filer.readdump(test_path)
        dict_ans = Filer.readdump(ans_path)
        filename = ans_path.replace('./files/rakuten_ichiba/dumpfile/ans/', '')
        filename = filename.replace('.dump', '')
        print counter
        counter += 1
        # テスト用ファイルの作成
        # １００単語以上ある文は除く
        list_sentence = []
        list_sep = []
        list_sep_all = []
        for row in list_test:
            list_sentence.append(row['sentence'])
            list_sep.append(row['sep_nj'])
            list_sep_all.append(row['sep_all'])
    
        # インスタンス化
        SM = SubModular(list_sep_all=list_sep_all,
                        list_sep=list_sep_all)
        for alpha in [0.01, 0.05, 0.1, 0.2]:
            for lamda in [0.5, 1, 5, 10]:
                for r in [0.1, 0.5, 1, 2]:
                    # 学習
                    n_cluster = len(list_sep_all)/5
                    SM.m_greedy(num_w=100, alpha=alpha, lamda=lamda, r=r, n_cluster=n_cluster)
                    list_C = SM.list_C
                    # Rougeを計算するようにbag_of_wordsを作成する
                    list_sm_test = [list_sep_all[row[0]] for row in list_C]
                    list_sm_ans = dict_ans['sep_all']
                    try:
                        R1 = Rouge_N.rouge(list_sm_test, list_sm_ans, N=1)
                    except:
                        R1 = 0
                    try:
                        R2 = Rouge_N.rouge(list_sm_test, list_sm_ans, N=2)
                    except:
                        R2 = 0
                    try:
                        R3 = Rouge_N.rouge(list_sm_test, list_sm_ans, N=3)
                    except:
                        R3 = 0
                    try:
                        R4 = Rouge_N.rouge(list_sm_test, list_sm_ans, N=4)
                    except:
                        R4 = 0
                    list_result = [[filename, alpha, lamda, r, R1, R2, R3, R4]]
                    Filer.writecsv(list_result, './files/rakuten_ichiba/result/SM/all/all_%s_%s_%s.csv'%(alpha, lamda, r))
    SlackBot.post_message(username=username,
                          text='%s号機、無事帰還しました！！' % username,
                          icon_emoji=':thumbsup:')
except:
    SlackBot.post_message(username=username,
                          text='エラー、下手くそ！！',
                          icon_emoji=':troll:')
    raise
```
