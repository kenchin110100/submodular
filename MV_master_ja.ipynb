```python
# coding: utf-8
"""
分散表現を用いた文書要約の精度の検証
tf-idfを用いて、文ベクトルを作成する
"""
from filer2.filer2 import Filer
from library.vector import Modified_Vector
from library.rouge import Rouge_N
from library.slackbot import SlackBot
import numpy as np
import glob
```

```python
start = 0
end = 35

list_test_path = glob.glob('./files/rakuten_ichiba/dumpfile/test/*')
list_ans_path = glob.glob('./files/rakuten_ichiba/dumpfile/ans/*')

list_test_path.sort()
list_ans_path.sort()

list_test_path = list_test_path[start:end]
list_ans_path = list_ans_path[start:end]

print len(list_test_path)
print len(list_ans_path)
```

```python
counter = 1

username = 'プログラムMV'

try:
    for test_path, ans_path in zip(list_test_path, list_ans_path):
        list_test = Filer.readdump(test_path)
        dict_ans = Filer.readdump(ans_path)
        filename = ans_path.replace('./files/rakuten_ichiba/tsvfile/ans/', '')
        filename = filename.replace('.dump', '')
        print counter
        counter += 1
        # テスト用ファイルの作成
        # １００単語以上ある文は除く
        list_sentence = []
        list_sep = []
        list_sep_all = []
        for row in list_test:
            list_sep.append(row['sep_nj'])
            list_sep_all.append(row['sep_all'])
            list_sentence.append(row['sentence'])

        # インスタンス化
        MV = Modified_Vector(list_sep_all=list_sep_all,
                             list_sep=list_sep,
                             dict_path = './files/rakuten_ichiba/w2v/model/model_cbow_s300_w5_m0/dict_word_vec_3')
    
        for scale in [0, 1, 2]:
            for r in [0.1, 0.5, 1, 2]:
    
                # 学習
                MV.m_greedy(num_w=100, r=r, scale=scale)
                # Rougeを計算するようにbag_of_wordsを作成する
                list_C = MV.list_C
                list_v_test = [list_sep_all[row[0]] for row in list_C]
                list_v_ans = dict_ans['sep_all']
                try:
                    R1 = Rouge_N.rouge(list_v_test, list_v_ans, N=1)
                except:
                    R1 = 0
                try:
                    R2 = Rouge_N.rouge(list_v_test, list_v_ans, N=2)
                except:
                    R2 = 0
                try:
                    R3 = Rouge_N.rouge(list_v_test, list_v_ans, N=3)
                except:
                    R3 = 0
                try:
                    R4 = Rouge_N.rouge(list_v_test, list_v_ans, N=4)
                except:
                    R4 = 0
                list_result = [[filename, scale, r, R1, R2, R3, R4]]
                Filer.writecsv(list_result, './files/rakuten_ichiba/result/MV/all/all_%s_%s.csv'%(scale, r))
                
    SlackBot.post_message(username=username,
                          text='%s号機、無事帰還しました！！' % username,
                          icon_emoji=':thumbsup:')
except:
    SlackBot.post_message(username=username,
                          text='エラー、下手くそ！！',
                          icon_emoji=':troll:')
    raise
```
